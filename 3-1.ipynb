```python
import pandas as pd
```

```python
df = pd.read_csv('data/Fish.csv')
```

```python
df.head()
```

```python
# 도미 데이터만 불러오기
cond = df['Species'] == 'Perch'
perch_data = df.loc[cond]
perch_data
```

```python
# 하나의 종의 수치 예측해볼 것
```

```python
import matplotlib.pyplot as plt

plt.scatter(perch_data['Length2'], perch_data['Weight'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = \
train_test_split(perch_data[['Length2']], perch_data[['Weight']])
```

```python
train_input.head()
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knr = KNeighborsRegressor()
```

```python
knr.fit(train_input, train_target)
```

```python
knr.score(test_input, test_target)
```

```python
# 결정계수
```

```python
# 예측하려는 값이 평균과 얼마나 오차가 있는지 확인
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
test_prediction = knr.predict(test_input)
# test_prediction
```

```python
mae = mean_absolute_error(test_target, test_prediction)
print(mae)
```

```python
# 과대적합 vs 과소적합
```

```python
knr.score(train_input, train_target) # 학습할 때 넣었던 데이터
```

```python
knr.score(test_input, test_target)
```

```python
# 학습한 점수가 96점, 채점할 때 점수가 99점이 나옴
# 왜 점수가 더 올라가는가?
# 과대적합과 과소적합이라고 함
# 보통 훈련세트 점수가 더 높음(그 데이터로 학습했기 때문)
# 둘 다 점수가 너무 낮다면 과소적합이라고 함
```

```python
knr.n_neighbors = 3 # 기본적으로 5로 설정되어 있음
```
